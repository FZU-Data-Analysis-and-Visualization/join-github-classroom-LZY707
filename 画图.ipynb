```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "SimHei"
plt.rcParams["axes.unicode_minus"] = False

# 月份1~12
x = np.arange(1, 13)
temp1 = [12, 15, 18, 22, 26, 29, 32, 33, 28, 24, 19, 14]  # A市气温
temp2 = [8, 11, 15, 19, 24, 27, 30, 31, 26, 21, 16, 10]   # B市气温

# 两市整体平均温度
all_temp = temp1 + temp2
avg_temp = np.mean(all_temp)

plt.figure(figsize=(11, 6))
# A城折线
plt.plot(x, temp1, color="#FF4500", lw=3, marker="o", ms=8, label="A城市", alpha=0.85)
# B城折线
plt.plot(x, temp2, color="#1E90FF", lw=3, marker="s", ms=8, label="B城市", alpha=0.85)
# 平均值水平虚线
plt.axhline(y=avg_temp, color="#222222", linestyle="--", lw=2, alpha=0.6, label=f'两市总平均气温:{avg_temp:.1f}℃')

# 图表美化
plt.title("AB两市全年月度平均气温对比", fontsize=16, pad=12)
plt.xlabel("月份", fontsize=13)
plt.ylabel("温度(℃)", fontsize=13)
plt.xticks(x)  
plt.grid(True, alpha=0.35, linestyle="-")
plt.legend(loc="upper right", fontsize=11)
plt.tight_layout()

plt.show()
```
