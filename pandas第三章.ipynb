### 一、基础创建与查看

```python
import pandas as pd

# 1. 用字典创建 DataFrame
data_dict = {
    'name': ['Tom', 'Jerry', 'Mike'],
    'score': [85, 92, 78],
    'gender': ['男', '女', '男']
}
df1 = pd.DataFrame(data_dict)

# 2. 查看 DataFrame 的前 1 行、形状、列名、数据类型
print("前1行数据：")
print(df1.head(1))
print("\n形状（行数, 列数）：")
print(df1.shape)
print("\n列名：")
print(df1.columns.tolist())
print("\n各列数据类型：")
print(df1.dtypes)

# 3. 用列表创建 DataFrame，列名设为 id、tag
data_list = [[1, "A"], [2, "B"], [3, "C"]]
df2 = pd.DataFrame(data_list, columns=['id', 'tag'])
print("\n列表创建的 DataFrame：")
print(df2)
```

### 二、列与行操作

```python
import pandas as pd

# 先创建好 df1
data_dict = {
    'name': ['Tom', 'Jerry', 'Mike'],
    'score': [85, 92, 78],
    'gender': ['男', '女', '男']
}
df1 = pd.DataFrame(data_dict)

# 1. 选取 score 和 gender 两列
df_step1 = df1[['score', 'gender']]
print("1. 选取 score 和 gender 两列：")
print(df_step1)

# 2. 选取索引 0、2 的行
df_step2 = df1.loc[[0, 2]]
print("\n2. 选取索引 0、2 的行：")
print(df_step2)

# 3. 筛选 score>80 的行
df_step3 = df1[df1['score'] > 80]
print("\n3. 筛选 score>80 的行：")
print(df_step3)

# 4. 新增一列 grade：score≥90 为优秀，否则为良好
df1['grade'] = df1['score'].apply(lambda x: '优秀' if x >= 90 else '良好')
print("\n4. 新增 grade 列后：")
print(df1)

# 5. 删除 gender 列
df_step5 = df1.drop(columns='gender')
print("\n5. 删除 gender 列后：")
print(df_step5)
```

### 三、缺失值

```python
import pandas as pd
import numpy as np

# 1. 创建含缺失值的 DataFrame
data = {"x": [1, None, 3], "y": [None, 5, 6]}
df = pd.DataFrame(data)
print("原始 DataFrame：")
print(df)

# 检测缺失值并统计每列缺失数量
print("\n缺失值检测结果（True 代表缺失）：")
print(df.isna())
print("\n每列缺失值数量：")
print(df.isna().sum())

# 2. 用 0 填充缺失值；删除含缺失值的行
df_fill = df.fillna(0)
print("\n用 0 填充缺失值后：")
print(df_fill)

df_drop = df.dropna()
print("\n删除含缺失值的行后：")
print(df_drop)
```

### 四、分组聚合

```python
import pandas as pd


data_dict = {
    'name': ['Tom', 'Jerry', 'Mike'],
    'score': [85, 92, 78],
    'gender': ['男', '女', '男']
}
df1 = pd.DataFrame(data_dict)

# 1. 按 gender 分组，求 score 的均值和最大值
group_result1 = df1.groupby('gender')['score'].agg(['mean', 'max'])
print("按 gender 分组，score 的均值和最大值：")
print(group_result1)

# 2. 按 gender 分组，统计每组人数
group_result2 = df1.groupby('gender').size()
print("\n按 gender 分组，每组人数：")
print(group_result2)
```

### 五、排序与去重

```python
import pandas as pd

# 最开始创建的原始数据
data_dict = {
    'name': ['Tom', 'Jerry', 'Mike'],
    'score': [85, 92, 78],
    'gender': ['男', '女', '男']
}
df1 = pd.DataFrame(data_dict)

# 1. 按 score 升序排序
df_sorted = df1.sort_values(by='score', ascending=True)
print("1. 按 score 升序排序后：")
print(df_sorted)

# 2. 添加一行重复数据（Tom,85,男）并去重
df1.loc[len(df1)] = ['Tom', 85, '男']
print("\n添加重复行后的数据：")
print(df1)

df_dedup = df1.drop_duplicates()
print("\n去重后的数据：")
print(df_dedup)
```
