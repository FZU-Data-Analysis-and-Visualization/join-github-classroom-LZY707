## 练习题4

```python
import numpy as np

arr = np.array([10,15,20,25,30,35])

# 花式索引
print(arr[[0, 2, 4]])

# 布尔索引
print(arr[arr > 20])
```

## 6

```python
import numpy as np

x = np.array([1, 4, 9, 16])
y = np.array([2, 1, 4, 5])

sqrt_x = np.sqrt(x)
print(sqrt_x)

result = np.maximum(sqrt_x, y)
print(result)
```

## 8

```python
import numpy as np

arr_x = np.array([1, 5, 7])
arr_y = np.array([2, 6, 8])
arr_con = np.array([True, False, True])

result = np.where(arr_con, arr_x, arr_y)
print(result)
```
