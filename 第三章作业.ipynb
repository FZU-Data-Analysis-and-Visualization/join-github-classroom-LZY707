### 二、列与行操作

```python
import pandas as pd

# 原始数据
data_dict = {
    'name': ['Tom', 'Jerry', 'Mike'],
    'score': [85, 92, 78],
    'gender': ['男', '女', '男']
}
df1 = pd.DataFrame(data_dict)

# 1. 选取 score 和 gender 两列
df1_1 = df1[['score', 'gender']]
print("1. 选取 score 和 gender 两列：")
print(df1_1)

# 2. 选取索引 0、2 的行
df1_2 = df1.loc[[0, 2]]
print("\n2. 选取索引 0、2 的行：")
print(df1_2)

# 3. 筛选 score>80 的行
df1_3 = df1[df1['score'] > 80]
print("\n3. 筛选 score>80 的行：")
print(df1_3)

# 4. 新增一列 grade：score≥90 为优秀，否则为良好
df1['grade'] = df1['score'].apply(lambda x: '优秀' if x >= 90 else '良好')
print("\n4. 新增 grade 列后：")
print(df1)

# 5. 删除 gender 列
df1_5 = df1.drop(columns='gender')
print("\n5. 删除 gender 列后：")
print(df1_5)
```

### 三、缺失值

```python
import pandas as pd

# 1. 创建含缺失值的 DataFrame
data = {"x": [1, None, 3], "y": [None, 5, 6]}
df = pd.DataFrame(data)
print("原始 DataFrame：")
print(df)

# 检测缺失值并统计每列缺失数量
print("\n缺失值检测（True 表示缺失）：")
print(df.isna())
print("\n每列缺失值数量：")
print(df.isna().sum())

# 2. 用 0 填充缺失值；删除含缺失值的行
df_fill = df.fillna(0)
print("\n用 0 填充缺失值后：")
print(df_fill)

df_drop = df.dropna()
print("\n删除含缺失值的行后：")
print(df_drop)
```

### 四、分组聚合

```python
import pandas as pd

# 原始数据
data_dict = {
    'name': ['Tom', 'Jerry', 'Mike'],
    'score': [85, 92, 78],
    'gender': ['男', '女', '男']
}
df1 = pd.DataFrame(data_dict)

# 1. 按 gender 分组，求 score 的均值和最大值
group_stats = df1.groupby('gender')['score'].agg(['mean', 'max'])
print("1. 按 gender 分组的 score 均值和最大值：")
print(group_stats)

# 2. 按 gender 分组，统计每组人数
group_count = df1.groupby('gender').size()
print("\n2. 按 gender 分组的每组人数：")
print(group_count)
```

### 五、排序与去重

```python
import pandas as pd

# 原始数据
data_dict = {
    'name': ['Tom', 'Jerry', 'Mike'],
    'score': [85, 92, 78],
    'gender': ['男', '女', '男']
}
df1 = pd.DataFrame(data_dict)

# 1. 按 score 升序排序
df_sorted = df1.sort_values(by='score', ascending=True)
print("1. 按 score 升序排序后：")
print(df_sorted)

# 2. 添加一行重复数据（Tom, 85, 男）并去重
df1.loc[len(df1)] = ['Tom', 85, '男']
print("\n添加重复行后的数据：")
print(df1)

df_deduplicated = df1.drop_duplicates()
print("\n去重后的数据：")
print(df_deduplicated)
```

```python
import pandas as pd

# 1. 横向合并（按 id 合并）
df1 = pd.DataFrame({"id": [1, 2], "name": ["X", "Y"]})
df2 = pd.DataFrame({"id": [1, 2], "age": [20, 30]})

df_merged = pd.merge(df1, df2, on="id")
print("1. 横向合并结果：")
print(df_merged)

# 2. 纵向拼接并重置索引
df_a = pd.DataFrame({"val": [1, 2]})
df_b = pd.DataFrame({"val": [3, 4]})

df_concat = pd.concat([df_a, df_b], axis=1)
print("\n2. 纵向拼接并重置索引结果：")
print(df_concat)
```

### 七、时间处理

```python
import pandas as pd

# 1. 创建时间列并转为 datetime 格式
dates = ["2025-01-01", "2025-01-02", "2025-01-03"]
df = pd.DataFrame({"date": dates})
df['date'] = pd.to_datetime(df['date'])
print("1. 转为 datetime 格式后：")
print(df)
print("\ndate 列的数据类型：", df['date'].dtype)

# 2. 提取时间列的月份、星期
df['month'] = df['date'].dt.month
df['weekday'] = df['date'].dt.weekday  # 0=周一, 1=周二, ..., 6=周日
df['weekday_name'] = df['date'].dt.day_name()  # 星期名称
print("\n2. 提取月份和星期后：")
print(df)
```

### 八、高级操作

```python
import pandas as pd

# 原始数据
data_dict = {
    'name': ['Tom', 'Jerry', 'Mike'],
    'score': [85, 92, 78],
    'gender': ['男', '女', '男']
}
df1 = pd.DataFrame(data_dict)

# 1. 对 score 列做标准化（(值 - 均值) / 标准差）
score_mean = df1['score'].mean()
score_std = df1['score'].std()
df1['score_std'] = (df1['score'] - score_mean) / score_std
print("1. 标准化后的 score：")
print(df1[['name', 'score', 'score_std']])

# 2. 用透视表：行是 gender，值是 score，聚合函数为中位数
pivot_table = pd.pivot_table(df1, index='gender', values='score', aggfunc='median')
print("\n2. 按 gender 分组的 score 中位数透视表：")
print(pivot_table)
```
